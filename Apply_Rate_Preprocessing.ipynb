```
import numpy as np
import pandas as pd 
import io
from matplotlib import pyplot as plt
```

```

# Impute the missing values of 'title_proximity_tfidf'
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
x_train = dataset['main_query_tfidf'][dataset['title_proximity_tfidf'].notnull()].values.reshape(-1,1)
y_train = dataset[dataset['title_proximity_tfidf'].notnull()]['title_proximity_tfidf'].values.reshape(-1,1)
x_test = dataset[dataset['title_proximity_tfidf'].isnull()]['main_query_tfidf'].values.reshape(-1,1)
y_test = dataset[dataset['title_proximity_tfidf'].isnull()]['title_proximity_tfidf'].values.reshape(-1,1)


reg.fit(x_train,y_train)

title_predicted = reg.predict(x_test)

title_predicted = title_predicted.flatten()


dataset.title_proximity_tfidf[dataset.title_proximity_tfidf.isnull()] = title_predicted
```

```
#Impute missing values of 'description_proximity_tfidf'
description = dataset['description_proximity_tfidf'].fillna((dataset['description_proximity_tfidf'].mean()))
dataset.description_proximity_tfidf[dataset.description_proximity_tfidf.isnull()] = description
```
